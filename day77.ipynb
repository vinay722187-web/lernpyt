```python
#overriding
class User1:
  def __init__(self,pen,pencil):
    self.pen=pen
    self.pencil=pencil
  def getpen(self):
    print('user1:',self.pen)

class User2(User1):
  def __init(self,pen ,pencil,eraser):
    # self.pen=pen
    # self.pencil=pencil
    self.eraser=eraser

    super().__init__(pen,pencil)

  def getpen(self):
    print('user2:',self.pen)

u=User2('blue','balck','doms')
u.getpen()
```

```python
#overriding
class User1:
  def __init__(self,pen,pencil):
    self.pen = pen
    self.pencil = pencil
  def getpen(self):
    print("user1: ",self.pen)

class User2(User1):
  def __init__(self,pen,pencil,earser):
    self.earser = earser
    super().__init__(pen,pencil)
  def getpen(self):
    print("user2: ",self.pen)

u = User2("blue","black","doms")
u.getpen()
print(u.pencil)
```

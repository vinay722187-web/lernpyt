```python
import pandas as pd
l=[1,2,3,4]
df=pd.Series(l)
print(df)

df=pd.Series(data=l,index=['vin','rr','ff','gg'])
print(df)

d={'vin':1,'pp':2,'oo':3,'ii':9}
df=pd.Series(data=d)
print(df)
```

```python
# pd is nickname of pandas
import pandas as pd
l = [10,20,30,40,50] # list
df = pd.Series(l) # use list in pd
print(df)


# example data
df = pd.Series(data=l)
print(df)

# example index
df = pd.Series(data=l , index=['nikhil','devendra','raj','hello','unknown'])
print(df)

# example key = index | value = value
d = {'nikhil':10,'devendra':20,'raj':30,'hello':40,'unknown':50}
df = pd.Series(data=d)
print(df)

# experiment
```

```python
#operation on colunm
import pandas as pd

data = {
    "Emp_ID": [101, 102, 103, 104, 105, 106],
    "Name": ["Amit", "Riya", "Raj", "Sara", "John", "Neha"],
    "Department": ["IT", "HR", "Finance", "IT", "Sales", "HR"],
    "Salary": [50000, 45000, 60000, 55000, 48000, 52000],
    "Experience": [2, 3, 5, 4, 1, 3]
}

df = pd.DataFrame(data)

#select colmn
print(df['Emp_ID'])

#select colmn store in variable
abc=df['Emp_ID']
print(abc)

# #colmn value change
# df['Emp_ID']=[1,2,3,4,5,6]
# print(df['Emp_ID'])

#coln value change
df['Emp_ID']=df['Emp_ID']-100
print(df['Emp_ID'])

#add new colmn
df['Amount'] = [0,0,0,0,0,0]
print(df)

#drop colmn
df.drop('Emp_ID',axis=1,inplace=True)
print(df)

#drop mult colmn
df.drop(['Name','Department'],axis=1,inplace=True)
```

```python
#operation on raws
import pandas as pd

data = {
    "Emp_ID": [101, 102, 103, 104, 105, 106],
    "Name": ["Amit", "Riya", "Raj", "Sara", "John", "Neha"],
    "Department": ["IT", "HR", "Finance", "IT", "Sales", "HR"],
    "Salary": [50000, 45000, 60000, 55000, 48000, 52000],
    "Experience": [2, 3, 5, 4, 1, 3]
}

df = pd.DataFrame(data)
print(df)

#loc 1st colmn not index and 2nd name
print(df.loc[1,"Name"])


#iloc row index >0 and colm index  >0
print(df.iloc[1,1])

#get mult raws
print(df.loc[[0,1,2]])

#get raws using iloc
print(df.iloc[[1,2,3]])

#mult rows and colmn in iloc
print(df.iloc[[0,1],[0,1]])

#mult rows and colmn in loc
print(df.loc[0:3,'Name':'Salary'])
```

```python
#FILTER DATA FRAME
import pandas as pd

data = {
    "Emp_ID": [101, 102, 103, 104, 105, 106],
    "Name": ["Amit", "Riya", "Raj", "Sara", "John", "Neha"],
    "Department": ["IT", "HR", "Finance", "IT", "Sales", "HR"],
    "Salary": [50000, 45000, 60000, 55000, 48000, 52000],
    "Experience": [2, 3, 5, 4, 1, 3]
}

df = pd.DataFrame(data)
print(df)

#filter one
print(df[df['Emp_ID']>104])

#filter 106
print(df[df['Emp_ID']==106])

#300000 to50000
# print(df[(condition one)&(condition 2)])
print(df[(df['Salary']>30000)&(df['Salary']<50000)])
```

```python
#SORTING
import pandas as pd

data = {
    "Emp_ID": [101, 102, 103, 104, 105, 106],
    "Name": ["Amit", "Riya", "Raj", "Sara", "John", "Neha"],
    "Department": ["IT", "HR", "Finance", "IT", "Sales", "HR"],
    "Salary": [50000, 45000, 60000, 55000, 48000, 52000],
    "Experience": [2, 3, 5, 4, 1, 3]
}

df = pd.DataFrame(data)
print(df)

#sorting ascending by default
print(df.sort_values('Salary'))

#sorting descending by default
print(df.sort_values('Salary',ascending =False))

#two colmn
print(df.sort_values(['Salary','Experience']))
```

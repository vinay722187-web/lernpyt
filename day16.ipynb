```python
#handle missing values
import numpy as np
import pandas as pd
data ={
    'experienced':[1,np.nan,3,4,5],
    'salary':[100,200,300,400,np.nan],
    'name':['hel','dhee','Nik','Raj','vin']
}
df=pd.DataFrame(data)
print(df)

#null values filter  find out
print(df.isnull())

#sum of null
print(df.isnull().sum())

#dropna  drop complete raw
df.dropna()                  #df.dropna(inplace=True)
print(df)

#fillna  null>fill
df.fillna(44444444)

print('PANDA FINISH')
```

```python
#matplotlib
import matplotlib.pyplot as plt     #used to visualisation in graph
data ={
    'experienced':[1,2,3,4,5],
    'salary':[100,200,300,400,500]
}

x=data['experienced']
y=data['salary']

#line chart
plt.figure(figsize=(4,3))
plt.plot(data['experienced'],data['salary'],color='darkblue',marker='*',markersize=50)

# plt.savefig(hello.jpg)
plt.show()
```

```python
#pie chrt
import matplotlib.pyplot as plt
x=['apple','bann','carrot','mango']
y=[3,4,5,6]
#z=['red','yellow','orange','green']
plt.figure(figsize=(4,5))
plt.pie(y,labels=x,autopct='%1.1f%%',colors=['red','yellow','orange','green'])    #colors=z
plt.show()
```

```python
#sub plot
import matplotlib.pyplot as plt

#pie chart
x=['apple','bann','carrot','mango']
y=[30,40,50,6]
plt.subplot(1,2,1)
plt.pie(y,labels=x)


# line chart
data={
    'experienced':[1,2,3,4,5],
    'salary':[100,200,300,400,500]
}

plt.subplot(2,2,2)
x=data['experienced']
y=data['salary']
plt.plot(x,y)
plt.tight_layout()
plt.show()
```
